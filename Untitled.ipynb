```python
import numpy as np
from datetime import datetime as dt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('mosquito_weather_wind.csv')
slim = pd.DataFrame()
slim = data[['date','number','lat','lon','Tmax','Tmin']]
slim = slim.reset_index(drop=False)
latmin = min(slim.lat)
latmax = max(slim.lat)
lonmin = min(slim.lon)
lonmax = max(slim.lon)
slim = slim.dropna()
X = slim[['lat','lon','Tmax','Tmin']]
y = slim.number
lin = LinearRegression().fit(X, y)
grd = np.indices((51,51)).T.reshape(-1,2)
grd = grd.astype(float)
grd[:,1] = (1/50.)*grd[:,1]*(lonmax - lonmin) + lonmin
grd[:,0] = (1/50.)*grd[:,0]*(latmax - latmin) + latmin
xx = np.linspace(lonmin,lonmax,51)
yy = np.linspace(latmin,latmax,51)

testX = pd.DataFrame({'lat':grd[:,0],'lon':grd[:,1]})
testX['Tmax'] = 73.63 
testX['Tmin'] = 65.18
TT = lin.predict(testX)
cont = pd.DataFrame({'num':TT})
```

```python
geo2 = [Point(xy) for xy in zip(testX["lon"],testX["lat"])]
geo_cont = gpd.GeoDataFrame(cont, crs = crs, geometry = geo2)
```
