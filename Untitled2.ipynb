```python
data = pd.read_csv('data/West_Nile_Virus__WNV__Mosquito_Test_Results.csv')
data = data.rename(columns={"TEST DATE": "date",
                            "NUMBER OF MOSQUITOES":"num",
                            "LATITUDE":"lat","LONGITUDE":"lon",
                            "TRAP":"trap",
                            "SEASON YEAR":"year",
                            "WEEK":"week",
                            "Census Tracts":"tract",
                            "Zip Codes":"zip",
                            "Community Areas":"community",
                            "Historical Wards 2003-2015":"ward"})
data['date'] = pd.to_datetime(data['date'])
data['date'] = data['date'].dt.date
data = data.dropna()
slim = pd.DataFrame()
slim = data[['trap','date','year','num','lat','lon']]
slim = slim.reset_index(drop=False)
slim['d_since']=0.
```

```python
by_trap = data.groupby('trap')
for trap, t in by_trap:
    T = t.sort_values(by='date', ascending=False)
    for i in range(len(T)-1):
        date = T.date.iloc[i]
#        dt = date - T.date.iloc[i+1]
        DT = (date - T.date.iloc[i+1]).total_seconds()/86400.
        ind = ((slim.date == date)&(slim.trap==trap))
        p = np.where(ind)[0]
        slim.at[p,'d_since']=DT
        break
```

```python
sns.relplot(x='d_since',y='num',kind= 'line',data=slim)
plt.xlim(0,20)
plt.show()
```
